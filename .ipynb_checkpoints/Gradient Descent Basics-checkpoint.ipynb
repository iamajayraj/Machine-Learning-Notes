```python
from sklearn.datasets import make_regression 
import numpy as np
```

```python
x,y=make_regression(n_samples=4,n_features=1,n_informative=1,n_targets=1,noise=80,random_state=13)
```

```python
x
```

```python
y
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
```

```python
reg.fit(x,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red')
```

```python
#Applying Gradient descent assuming slope is m=78.35063668
#lets start with b=100 and then narrow down the value upto 
y_pred=((78.35063668 * x) + 100).reshape(4)
y_pred
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red',label="OLS line")
plt.plot(x,y_pred,color='black',label="b=100 line")
plt.legend()
```

```python
b=100
b
```

```python
m=78.35063668
m
```

```python
lr=0.1
lr
```

```python
slope_b=-2 * np.sum(y - m*x.ravel() - b)
slope_b
```

```python
sum(y - m*x.ravel() - b)
```

```python
np.sum(y - m*x.ravel() - b)
```

```python
b = b - (slope_b * lr)
b
```

```python
y_pred1=(m*x+b).reshape(4)
y_pred1
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red',label="OLS line")
plt.plot(x,y_pred,color='grey',label="b=100 line")
plt.plot(x,y_pred1,color='black',label="b line")
plt.legend()
```

```python
slope_b = -2 * np.sum(y - m*x.ravel() - b) 
slope_b
```

```python
b= b - (slope_b * lr)
b
```

```python
y_pred2=(m * x + b).reshape(4)
y_pred2
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red',label="OLS line")
plt.plot(x,y_pred,color='grey',label="b=100 line")
plt.plot(x,y_pred1,color='black',label="b line")
plt.plot(x,y_pred2,color='blue',label="b line")

plt.legend()
```

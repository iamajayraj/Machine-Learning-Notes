```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
```

```python
df = pd.read_csv("E:\Datasets\session 24\iris.csv")
df
```

```python
df = df.drop('Id',axis=1)
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
encoder= LabelEncoder()
```

```python
x=df.iloc[:,:4]
```

```python
x
```

```python
y=df.iloc[:,4]
```

```python
y
```

```python
y= encoder.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=5)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
nb= GaussianNB()
```

```python
nb.fit(x_train,y_train)
```

```python
y_pred = nb.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
```

```python
x_train
```

```python
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
scaled_nb = GaussianNB()
```

```python
scaled_nb.fit(x_train,y_train)
```

```python
y_pred_scaled = scaled_nb.predict(x_test)
```

```python
accuracy_score(y_test,y_pred_scaled)
```

## ##Conclusion

Clearly there is no change in accuracy of results before and after scaling.
The reason is that naive bayes simply calculates the probablity. So it doesnt make any difference.
